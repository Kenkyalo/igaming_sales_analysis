```python
df['Calculated_Global_Sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
(df['Global_Sales'] == df['Calculated_Global_Sales']).value_counts()
```

which game genre sells best globally


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Genre', y='Global_Sales', estimator='sum', ci=None)
plt.title('Total Global Sales by Genre')
plt.xticks(rotation=45)
plt.ylabel('Total Global Sales (Millions)')
plt.tight_layout()
plt.show()
```

Which Region Contributes Most to Sales?

```python
sales_by_region = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(8,5))
sales_by_region.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Region')
plt.ylabel('Sales (Millions)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

 Sales Over Time (by Year)

```python
df['Year_Numeric'] = pd.to_numeric(df['Year'], errors='coerce')
df_cleaned = df.dropna(subset=['Year_Numeric'])
df_by_year = df_cleaned.groupby('Year_Numeric')['Global_Sales'].sum()

plt.figure(figsize=(12,5))
df_by_year.plot(kind='line', marker='o')
plt.title('Global Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales (Millions)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
